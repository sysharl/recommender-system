### Singular Value Decomposition Recommender

```python
import sklearn
from sklearn.decomposition import TruncatedSVD
```

```python
columns = ['user_id','item_id','rating','timestamp']
frame = pd.read_csv('data/u.data',sep='\t', names=columns)
frame.head()
```

```python
movies_columns = ['item_id','movie_title','release date','video release date','IMDb URL',
                  'unknown','Action','Adventure','Animation','Childrens','Comedy','Crime',
                  'Documentary','Drama','Fantasy','Film-Noir','Horror','Musical','Mystery',
                  'Romance','Sci-Fi','Thriller','War','Western']
movies = pd.read_csv('data/u.item',sep='|', names=movies_columns, encoding='latin-1')
movie_names = movies.loc[:,['movie_title','item_id']]
movie_names.head()
```

```python
combined_movies_data = pd.merge(frame,movie_names,on='item_id')
combined_movies_data.head()
```

```python
combined_movies_data.groupby('item_id')['rating'].count().sort_values(ascending=False).head()
```

```python
popular_rated = combined_movies_data['item_id']==50
combined_movies_data[popular_rated]['movie_title'].unique()
```

#### Building Utility Matrix

```python
rating_crosstab = combined_movies_data.pivot_table(values ='rating', index='user_id',columns='movie_title', fill_value=0)
rating_crosstab.head()
```

```python
combined_movies_data[(combined_movies_data['user_id']==1) & (combined_movies_data['movie_title']=='Young Frankenstein (1974)')]
```

#### Transposing the Matrix

```python
rating_crosstab.shape
```

```python
X = rating_crosstab.values.T
X.shape
```

#### Decomposing the Matrix

```python
SVD = TruncatedSVD(n_components=12, random_state=17) #Dimensionality reduction using truncated SVD
resultant_matrix = SVD.fit_transform(X)
resultant_matrix.shape
```

#### Generating a Correlation Matrix

```python
corr_mat = np.corrcoef(resultant_matrix) #generate correlation matrix
corr_mat.shape
```

#### Isolate Star Wars From the Correlation Matrix

```python
movie_names = rating_crosstab.columns
movie_list = list(movie_names)
star_wars_index= movie_list.index('Star Wars (1977)')
```

```python
corr_star = corr_mat[star_wars_index]
corr_star
```

```python
movie_names[(corr_star<1.0) & (corr_star>0.9)]
```
